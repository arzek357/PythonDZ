# Краткий экскурс по стандартным библиотекам <a class="anchor" id="standart_libraries_briefing"></a>

Модули в python импортуруются с помощью выражения **import modulename**. После этого модно воспользоваться методами из импортируемого модуля с помощью точечной нотации как **modulename.methodname**. Имена модуоей при импорте можно заменять на более удобные, например, **import modulename sa mn**. В этом случае обращение к методу **methodname** будет выглядеть как **mn.methodname**.

В данном примере импортируется весь модуль и мы имеем доступ ко всем его методам. Если же нам нужа какая-то небольшая часть моуля, то можно импортировать только необходимые части **from modulename import methodname1, methodname2, methodname3**

* **модуль argparse**

Предоставляет набор функций для обработки аргументов командной строки.

Предоставляет набор функций для работы с регулярными выражениям (regular expressions)

# Установка сторонних модулей <a class="anchor" id="thirdparty_libraries"></a>

Также вы можете импортировать сторонние модули. Для этого их необходимо скачать и установить. Самый распространенный способ это сделать - воспользоваться модулем **pip**(preferred installer program). Начиная с версии 3.4. он включен в стандартные модули python. Если **pip** не установлен по умолчанию, можно попробовать прописать команду **python -m ensurepip --default-pip** (один из методов).



Для установки модулей вам необходимо прописать следующую команду из командной строки:

**python -m pip install SomePackage**

или

**pip install SomePackage**

Этой командой вы установите самую последнюю версию модуля SomePackage. Возможно вам придется прописать путь к python или pip переменных окружения (system variables или user variables для Windows).

Если необходимо поставить конкретную версию модуля можно видоизменить команду:

-Конкретная версия **python -m pip install SomePackage==1.0.4**
-Минимальная версия **python -m pip install SomePackage>=1.0.4**

Если модуль уже установлен, то повторная установка не окажет никакого эффекта. Обновить модуль можно так:

**python -m pip install --upgrade SomePackage**

Для разных версий python:
* python2   -m pip install SomePackage  # default Python 2
* python2.7 -m pip install SomePackage  # specifically Python 2.7
* python3   -m pip install SomePackage  # default Python 3
* python3.4 -m pip install SomePackage  # specifically Python 3.4

Для установки пакета только для текущего пользователя воспользуйтесь аргументом **--user**

**python -m pip install --user SomePackage**

Здесь мы описали самый минимум возможностей pip, который вам может пригодится для быстрого старта и его использования

# Virtual enviroment <a class="anchor" id="venv"></a>

Приложения Python часто используют сторонние пакеты и модули. Разным приложениям могут требоваться различные версии модулей по каким либо причинам.

Это означает, что одна установка модуля может не соответствовать требованиям каждого приложения. Если приложению A требуется версия 1.0 определенного модуля, а приложению B требуется версия 2.0, то мы не можем с помощью pip удовлетворить этому условию и установка версии 1.0 или 2.0 приведет к невозможности запуска другого приложения.

Решением этой проблемы является создание виртуального окружения, независимого дерева директорий, содержащего свои собственные установленные пакеты python, которые никак не будут влиять на основную версию python или другие виртуальные окружения.

Для создания виртуального окружения перейдите в директорию, где вы хотите создать и ранить это окружение, а затем пропишите команду **python3 -m venv env_name**.

Таким образом было создано виртуальное окружение **env_name** в текущей директории. Оно включает в себя интерпретатор и стандартные библиотеки.

Теперь необходимо активировать это окружение:

source env_name/bin/activate

После этого, командная строка изменится и будет показывать, какое окружение вы импользуете (для деактивации окружения достаточно прописать команду **deactivate**).

Теперь находясь в виртуальном окружении мы можем ставить необходимые пакеты с помощью **pip** как было описано выше
